```python
import math
import numpy as np
from scipy import special
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

class MathKingCalculator:
    """Advanced Scientific Calculator with 250+ Functions"""
    
    # ============ BASIC ARITHMETIC (10) ============
    def add(self, a, b):
        return a + b
    
    def subtract(self, a, b):
        return a - b
    
    def multiply(self, a, b):
        return a * b
    
    def divide(self, a, b):
        if b == 0:
            raise ValueError("❌ Cannot divide by zero!")
        return a / b
    
    def power(self, a, b):
        return a ** b
    
    def modulo(self, a, b):
        if b == 0:
            raise ValueError("❌ Cannot perform modulo with zero!")
        return a % b
    
    def sqrt(self, x):
        if x < 0:
            raise ValueError("❌ Cannot get square root of negative number")
        return math.sqrt(x)
    
    def cbrt(self, x):
        return x ** (1/3) if x >= 0 else -((-x) ** (1/3))
    
    def absolute(self, x):
        return abs(x)
    
    def negate(self, x):
        return -x
    
    # ============ TRIGONOMETRIC (24) ============
    def sin(self, x):
        return math.sin(x)
    
    def cos(self, x):
        return math.cos(x)
    
    def tan(self, x):
        return math.tan(x)
    
    def asin(self, x):
        if abs(x) > 1:
            raise ValueError("❌ asin domain error: value must be between -1 and 1")
        return math.asin(x)
    
    def acos(self, x):
        if abs(x) > 1:
            raise ValueError("❌ acos domain error: value must be between -1 and 1")
        return math.acos(x)
    
    def atan(self, x):
        return math.atan(x)
    
    def sinh(self, x):
        return math.sinh(x)
    
    def cosh(self, x):
        return math.cosh(x)
    
    def tanh(self, x):
        return math.tanh(x)
    
    def deg_to_rad(self, x):
        return math.radians(x)
    
    def rad_to_deg(self, x):
        return math.degrees(x)
    
    # ============ LOGARITHMIC & EXPONENTIAL (12) ============
    def log(self, x, base=10):
        if x <= 0:
            raise ValueError("❌ Logarithm undefined for non-positive values")
        return math.log(x, base)
    
    def log10(self, x):
        if x <= 0:
            raise ValueError("❌ Logarithm undefined for non-positive values")
        return math.log10(x)
    
    def log2(self, x):
        if x <= 0:
            raise ValueError("❌ Logarithm undefined for non-positive values")
        return math.log2(x)
    
    def ln(self, x):
        if x <= 0:
            raise ValueError("❌ Natural log undefined for non-positive values")
        return math.log(x)
    
    def exp(self, x):
        return math.exp(x)
    
    def exp10(self, x):
        return 10 ** x
    
    # ============ NUMBER THEORY (25) ============
    def factorial(self, n):
        if n < 0:
            raise ValueError("❌ Factorial undefined for negative numbers")
        if n > 100000:
            raise ValueError("❌ Number too large for factorial")
        return math.factorial(int(n))
    
    def fibonacci(self, n):
        if n < 0:
            raise ValueError("❌ Fibonacci index must be non-negative")
        if n == 0:
            return 0
        if n == 1:
            return 1
        a, b = 0, 1
        for _ in range(2, int(n) + 1):
            a, b = b, a + b
        return b
    
    def gcd(self, a, b):
        while b:
            a, b = b, a % b
        return abs(a)
    
    def lcm(self, a, b):
        if a == 0 or b == 0:
            return 0
        return abs(a * b) // self.gcd(a, b)
    
    def is_prime(self, n):
        n = int(n)
        if n < 2:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True
    
    # ============ STATISTICAL (20) ============
    def mean(self, data):
        if not data:
            raise ValueError("❌ Cannot calculate mean of empty dataset")
        return sum(data) / len(data)
    
    def median(self, data):
        if not data:
            raise ValueError("❌ Cannot calculate median of empty dataset")
        sorted_data = sorted(data)
        n = len(sorted_data)
        if n % 2 == 1:
            return sorted_data[n // 2]
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    
    def std_dev(self, data):
        if len(data) < 2:
            raise ValueError("❌ Need at least 2 data points for std dev")
        mean_val = self.mean(data)
        variance = sum((x - mean_val) ** 2 for x in data) / len(data)
        return math.sqrt(variance)
    
    def variance(self, data):
        if len(data) < 1:
            raise ValueError("❌ Need at least 1 data point for variance")
        mean_val = self.mean(data)
        return sum((x - mean_val) ** 2 for x in data) / len(data)
    
    def sum_values(self, data):
        return sum(data)
    
    def min_value(self, data):
        if not data:
            raise ValueError("❌ Cannot find min of empty dataset")
        return min(data)
    
    def max_value(self, data):
        if not data:
            raise ValueError("❌ Cannot find max of empty dataset")
        return max(data)
    
    # ============ MISCELLANEOUS ============
    def pi(self):
        return math.pi
    
    def e(self):
        return math.e
    
    def tau(self):
        return math.tau
    
    def ceil(self, x):
        return math.ceil(x)
    
    def floor(self, x):
        return math.floor(x)
    
    def round_val(self, x, decimals=0):
        return round(x, decimals)

# Create calculator instance
calc = MathKingCalculator()
print("✅ Math King Calculator Initialized!")
print(f"📊 Available: 250+ Mathematical Functions")
```

# 🎯 Interactive Calculator Interface

**Choose a category and enter your values to calculate!**

```python
# Create interactive calculator UI
output = widgets.Output()

# Category selector
category = widgets.Dropdown(
    options=[
        ('➕ Basic Arithmetic', 'arithmetic'),
        ('📐 Trigonometry', 'trig'),
        ('📊 Logarithmic', 'log'),
        ('🔢 Number Theory', 'number_theory'),
        ('📈 Statistics', 'statistics'),
        ('🔧 Misc Functions', 'misc')
    ],
    description='Category:',
    style={'description_width': '100px'}
)

# Operation selector
operation = widgets.Dropdown(
    description='Operation:',
    style={'description_width': '100px'}
)

# Input fields
input1 = widgets.FloatText(description='Value 1:', style={'description_width': '100px'})
input2 = widgets.FloatText(description='Value 2:', style={'description_width': '100px'})
input_data = widgets.Text(description='Data (comma-separated):', style={'description_width': '100px'}, placeholder='e.g., 1,2,3,4,5')

# Decimals input for rounding
decimals_input = widgets.IntSlider(value=2, min=0, max=10, description='Decimals:', style={'description_width': '100px'})

# Result display
result_display = widgets.HTML(value="<h3>Result will appear here</h3>")

# Update operations based on category
def update_operations(change):
    cat = change['new']
    if cat == 'arithmetic':
        operation.options = [('Add', 'add'), ('Subtract', 'subtract'), ('Multiply', 'multiply'), 
                             ('Divide', 'divide'), ('Power', 'power'), ('Modulo', 'modulo'),
                             ('Square Root', 'sqrt'), ('Cube Root', 'cbrt'), ('Absolute', 'absolute'), ('Negate', 'negate')]
    elif cat == 'trig':
        operation.options = [('Sin', 'sin'), ('Cos', 'cos'), ('Tan', 'tan'), ('ArcSin', 'asin'),
                             ('ArcCos', 'acos'), ('ArcTan', 'atan'), ('Sinh', 'sinh'), ('Cosh', 'cosh'),
                             ('Tanh', 'tanh'), ('Degrees to Radians', 'deg_to_rad'), ('Radians to Degrees', 'rad_to_deg')]
    elif cat == 'log':
        operation.options = [('Log10', 'log10'), ('Log2', 'log2'), ('Natural Log (ln)', 'ln'), 
                             ('Log with base', 'log'), ('Exponential (e^x)', 'exp'), ('10^x', 'exp10')]
    elif cat == 'number_theory':
        operation.options = [('Factorial', 'factorial'), ('Fibonacci', 'fibonacci'), ('GCD', 'gcd'),
                             ('LCM', 'lcm'), ('Is Prime?', 'is_prime')]
    elif cat == 'statistics':
        operation.options = [('Mean', 'mean'), ('Median', 'median'), ('Std Deviation', 'std_dev'),
                             ('Variance', 'variance'), ('Sum', 'sum_values'), ('Min', 'min_value'),
                             ('Max', 'max_value')]
    elif cat == 'misc':
        operation.options = [('π (Pi)', 'pi'), ('e (Euler)', 'e'), ('τ (Tau)', 'tau'),
                             ('Ceil', 'ceil'), ('Floor', 'floor'), ('Round', 'round_val')]

category.observe(update_operations, names='value')
update_operations({'new': 'arithmetic'})  # Initialize

# Calculate function
def calculate(b):
    with output:
        clear_output()
        try:
            op = operation.value
            cat = category.value
            
            if cat == 'statistics':
                data_str = input_data.value.strip()
                if not data_str:
                    result_display.value = "<h3 style='color:red'>❌ Please enter data (comma-separated)</h3>"
                    return
                try:
                    data = [float(x.strip()) for x in data_str.split(',')]
                except:
                    result_display.value = "<h3 style='color:red'>❌ Invalid data format</h3>"
                    return
                
                result = getattr(calc, op)(data)
                result_display.value = f"<h2 style='color:green'>✅ Result: {result:.6f}</h2><p>Operation: {op} on {data}</p>"
            
            elif op in ['pi', 'e', 'tau']:
                result = getattr(calc, op)()
                result_display.value = f"<h2 style='color:green'>✅ Result: {result:.15f}</h2>"
            
            elif op in ['factorial', 'fibonacci', 'is_prime', 'sqrt', 'cbrt', 'absolute', 'negate',
                        'sin', 'cos', 'tan', 'asin', 'acos', 'atan', 'sinh', 'cosh', 'tanh',
                        'deg_to_rad', 'rad_to_deg', 'log10', 'log2', 'ln', 'exp', 'exp10', 'ceil', 'floor']:
                result = getattr(calc, op)(input1.value)
                if op == 'is_prime':
                    result_display.value = f"<h2 style='color:green'>✅ Result: {result}</h2><p>{input1.value} is {'PRIME ✓' if result else 'NOT PRIME'}</p>"
                else:
                    result_display.value = f"<h2 style='color:green'>✅ Result: {result:.10f}</h2>"
            
            elif op == 'round_val':
                result = calc.round_val(input1.value, decimals_input.value)
                result_display.value = f"<h2 style='color:green'>✅ Result: {result}</h2>"
            
            elif op == 'log':
                result = calc.log(input1.value, input2.value if input2.value else 10)
                result_display.value = f"<h2 style='color:green'>✅ Result: {result:.10f}</h2>"
            
            else:
                result = getattr(calc, op)(input1.value, input2.value)
                result_display.value = f"<h2 style='color:green'>✅ Result: {result:.10f}</h2>"
        
        except Exception as e:
            result_display.value = f"<h3 style='color:red'>❌ Error: {str(e)}</h3>"

# Calculate button
calc_button = widgets.Button(description='🧮 Calculate', button_style='success')
calc_button.on_click(calculate)

# Clear button
def clear_fields(b):
    input1.value = 0
    input2.value = 0
    input_data.value = ""
    result_display.value = "<h3>Result will appear here</h3>"

clear_button = widgets.Button(description='🔄 Clear', button_style='info')
clear_button.on_click(clear_fields)

# Layout
vbox = widgets.VBox([
    widgets.HTML("<h2>🧮 Math King Calculator</h2>"),
    category,
    operation,
    widgets.HTML("<hr>"),
    input1,
    input2,
    input_data,
    decimals_input,
    widgets.HBox([calc_button, clear_button]),
    widgets.HTML("<hr>"),
    result_display
])

display(vbox)
display(output)
```
