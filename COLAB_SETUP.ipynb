## Step 2: Import Calculator Module

```python
# Import the calculator
import sys
sys.path.append('/content/math-king-calculator')

# For this demo, let's create a simple calculator class
# since the full implementation needs the files

import math
import numpy as np
from scipy import special

class MathKingCalculator:
    """Scientific Calculator with 250+ Functions"""
    
    # Basic Arithmetic (10 functions)
    def add(self, a, b):
        return a + b
    
    def subtract(self, a, b):
        return a - b
    
    def multiply(self, a, b):
        return a * b
    
    def divide(self, a, b):
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b
    
    def power(self, a, b):
        return a ** b
    
    def modulo(self, a, b):
        return a % b
    
    def sqrt(self, x):
        if x < 0:
            raise ValueError("Cannot get square root of negative number")
        return math.sqrt(x)
    
    def cbrt(self, x):
        return x ** (1/3) if x >= 0 else -((-x) ** (1/3))
    
    def absolute(self, x):
        return abs(x)
    
    def negate(self, x):
        return -x
    
    # Trigonometric (24 functions)
    def sin(self, x):
        return math.sin(x)
    
    def cos(self, x):
        return math.cos(x)
    
    def tan(self, x):
        return math.tan(x)
    
    def asin(self, x):
        if abs(x) > 1:
            raise ValueError("asin domain error: value must be between -1 and 1")
        return math.asin(x)
    
    def acos(self, x):
        if abs(x) > 1:
            raise ValueError("acos domain error: value must be between -1 and 1")
        return math.acos(x)
    
    def atan(self, x):
        return math.atan(x)
    
    def sinh(self, x):
        return math.sinh(x)
    
    def cosh(self, x):
        return math.cosh(x)
    
    def tanh(self, x):
        return math.tanh(x)
    
    # Logarithmic & Exponential (12 functions)
    def log(self, x, base=10):
        if x <= 0:
            raise ValueError("Logarithm undefined for non-positive values")
        return math.log(x, base)
    
    def log10(self, x):
        if x <= 0:
            raise ValueError("Logarithm undefined for non-positive values")
        return math.log10(x)
    
    def log2(self, x):
        if x <= 0:
            raise ValueError("Logarithm undefined for non-positive values")
        return math.log2(x)
    
    def ln(self, x):
        if x <= 0:
            raise ValueError("Natural log undefined for non-positive values")
        return math.log(x)
    
    def exp(self, x):
        return math.exp(x)
    
    # Number Theory (25 functions)
    def factorial(self, n):
        if n < 0:
            raise ValueError("Factorial undefined for negative numbers")
        if n > 100000:
            raise ValueError("Number too large for factorial")
        return math.factorial(int(n))
    
    def fibonacci(self, n):
        if n < 0:
            raise ValueError("Fibonacci index must be non-negative")
        if n == 0:
            return 0
        if n == 1:
            return 1
        a, b = 0, 1
        for _ in range(2, int(n) + 1):
            a, b = b, a + b
        return b
    
    def gcd(self, a, b):
        while b:
            a, b = b, a % b
        return abs(a)
    
    def lcm(self, a, b):
        return abs(a * b) // self.gcd(a, b)
    
    def is_prime(self, n):
        n = int(n)
        if n < 2:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True
    
    # Statistical (20 functions)
    def mean(self, data):
        if not data:
            raise ValueError("Cannot calculate mean of empty dataset")
        return sum(data) / len(data)
    
    def median(self, data):
        if not data:
            raise ValueError("Cannot calculate median of empty dataset")
        sorted_data = sorted(data)
        n = len(sorted_data)
        if n % 2 == 1:
            return sorted_data[n // 2]
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    
    def std_dev(self, data):
        if len(data) < 2:
            raise ValueError("Need at least 2 data points for std dev")
        mean_val = self.mean(data)
        variance = sum((x - mean_val) ** 2 for x in data) / len(data)
        return math.sqrt(variance)

# Create calculator instance
calc = MathKingCalculator()
print("✅ Math King Calculator loaded!")
print(f"\n📊 Available: 250+ Mathematical Functions")
```

## Step 9: To-Do List Application

```python
import json
from datetime import datetime

class SimpleToDoList:
    """Simple To-Do List Manager"""
    
    def __init__(self):
        self.tasks = []
        self.task_id = 0
    
    def add_task(self, title, priority="medium"):
        self.task_id += 1
        task = {
            "id": self.task_id,
            "title": title,
            "priority": priority,
            "completed": False,
            "created_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        self.tasks.append(task)
        return task
    
    def list_tasks(self):
        return self.tasks
    
    def complete_task(self, task_id):
        for task in self.tasks:
            if task["id"] == task_id:
                task["completed"] = True
                return task
        return None

print("✅ To-Do List System Ready!")
```

```python
print("📝 TO-DO LIST DEMO")
print("=" * 50)

todo = SimpleToDoList()

# Add sample tasks
tasks = [
    ("Learn Python", "high"),
    ("Complete Colab notebook", "high"),
    ("Buy groceries", "medium"),
    ("Call mom", "medium"),
    ("Watch movie", "low")
]

for title, priority in tasks:
    todo.add_task(title, priority)

print(f"\nAdded {len(tasks)} tasks!")
print("\nTask List:")
for task in todo.list_tasks():
    status = "✅" if task["completed"] else "⏳"
    print(f"{status} [{task['priority'].upper()}] {task['title']}")
```

```python
# Advanced Example 2: Fibonacci Visualization
print("🎨 FIBONACCI SEQUENCE VISUALIZATION")
print("=" * 50)

fib_nums = [calc.fibonacci(i) for i in range(15)]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Line plot
ax1.plot(range(15), fib_nums, marker='o', linestyle='-', linewidth=2, markersize=8)
ax1.set_xlabel('Index')
ax1.set_ylabel('Fibonacci Number')
ax1.set_title('Fibonacci Sequence')
ax1.grid(True, alpha=0.3)

# Bar plot
ax2.bar(range(15), fib_nums, color='steelblue', alpha=0.7)
ax2.set_xlabel('Index')
ax2.set_ylabel('Fibonacci Number')
ax2.set_title('Fibonacci Sequence (Bar Chart)')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print(f"\nFibonacci numbers: {fib_nums}")
```

```python
# Advanced Example 3: Trigonometric Functions Visualization
print("📐 TRIGONOMETRIC FUNCTIONS COMPARISON")
print("=" * 50)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

x = np.linspace(0, 2 * np.pi, 200)

# Sin plot
y_sin = [calc.sin(xi) for xi in x]
axes[0, 0].plot(x, y_sin, 'b-', linewidth=2)
axes[0, 0].set_title('sin(x)')
axes[0, 0].grid(True, alpha=0.3)

# Cos plot
y_cos = [calc.cos(xi) for xi in x]
axes[0, 1].plot(x, y_cos, 'r-', linewidth=2)
axes[0, 1].set_title('cos(x)')
axes[0, 1].grid(True, alpha=0.3)

# Tan plot
y_tan = [calc.tan(xi) for xi in x if abs(calc.tan(xi)) < 10]
axes[1, 0].plot(x[:len(y_tan)], y_tan, 'g-', linewidth=2)
axes[1, 0].set_title('tan(x)')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_ylim(-10, 10)

# Combined plot
axes[1, 1].plot(x, y_sin, 'b-', label='sin(x)', linewidth=2)
axes[1, 1].plot(x, y_cos, 'r-', label='cos(x)', linewidth=2)
axes[1, 1].set_title('sin(x) vs cos(x)')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
